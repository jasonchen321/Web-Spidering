```python
import requests
import pandas as pd
import json
```

```python
url = requests.get('https://www.maximsmx.com.hk/branch_list')
```

```python
url = url.json()
```

```python
url['1']['branches']
```

```python
name = []
addr = []
tel = []
X = []
Y = []
opening = []
for i in range(1,4):
    for j in range(len(url[str(i)]['branches'])):
        name.append(url[str(i)]['branches'][j]['name'])
        addr.append(url[str(i)]['branches'][j]['address'])
        tel.append(url[str(i)]['branches'][j]['tel'])
        X.append(url[str(i)]['branches'][j]['position']['lng'])
        Y.append(url[str(i)]['branches'][j]['position']['lat'])
        opening.append(url[str(i)]['branches'][j]['opening_info'])
```

```python
df = pd.DataFrame({'Name':name,'Address':addr,'Tel':tel,'Lon':X,'Lat':Y,'Opening':opening}, columns=['Name','Address','Tel','Lon','Lat','Opening'])
```
