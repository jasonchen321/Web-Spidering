```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import xmltodict
import xml.etree.ElementTree as ET
import json
```

```python
url = requests.get("https://www.icbcasia.com/ICBC/%e6%b5%b7%e5%a4%96%e5%88%86%e8%a1%8c/%e5%b7%a5%e9%93%b6%e4%ba%9a%e6%b4%b2/TC/%e9%97%9c%e6%96%bc%e6%88%91%e5%80%91/%e8%81%af%e7%b5%a1%e6%88%91%e5%80%91/%e5%88%86%e8%a1%8c%e7%b6%b2%e7%b5%a1/default.htm")
```

```python
url.encoding=('utf8')
```

```python
url.text
```

```python
soup = BeautifulSoup(url.text, 'html.parser')
```

```python
s = str(soup)
```

```python
ss = s.find("var hongKongData    = ")
print(ss)
e = s.find("// ntdata end")
print(e)
rep = {'[':"",']':"",'<p>':"","\r\n        ":"","\r\n            ":"","\',":"@","\'":"@"}
```

```python
cleandoc = s.replace(s[0:ss+len("var hongKongData    = ")],"")
cleandoc = cleandoc.replace(s[e+len("// ntdata end"):],"")
for a,b in rep.items():
    cleandoc = cleandoc.replace(a,b)
```

```python
cleandoc = re.sub(r'<[^<>]*>','', cleandoc)
```

```python
doclist = cleandoc.split("@")
```

```python
del doclist[0]
```

```python
llist = [doclist[i:i+6] for i in range(0,len(doclist),6)]
```

```python
name = []
address = []
coor = []
phone = []
for j in range(len(llist)):
    name.append(llist[j][0])
    coor.append(llist[j][1])
    phoneindex = llist[j][2].find("電話:")
    #addrsign = ["號","鋪"]
    addrindex = len(llist[j][2])
    #print(phoneindex)
 
    if phoneindex != -1:
        phonestart = phoneindex+len("電話:")
        phone.append(llist[j][2][phonestart:phonestart+14])
    else:    
        phone.append(" ")
    
    
    if phoneindex != -1:
        address.append(llist[j][2][:phoneindex-1])
    elif addrindex != -1:
        addrc = llist[j][2].replace(" ","")
        address.append(addrc[0:phoneindex-1])
    else:
        address.append(" ")
```

```python
coorclean = []
coorclean1 = []
for k in range(len(coor)):
    d = coor[k].replace(' ','')
    coorclean.append(d)
for l in range(len(coorclean)):
    llen = len(coorclean[l])
    a = coorclean[l][:llen-1]
    coorclean1.append(a)
```

```python
len(name),len(address),len(phone),len(coorclean1)
```

```python
df = pd.DataFrame({'Name':name,'Address':address,'Phone':phone,'Coor':coorclean1},columns=['Name','Address','Phone','Coor'])
```
