```python
from slimit import ast
from slimit.parser import Parser
from slimit.visitors import nodevisitor
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html = requests.get("https://www.dbs.com.hk/iwov-resources/scripts/maps/data-zh.js")
```

```python
html.text
```

```python
parser = Parser()
tree = parser.parse(html.text)
```

```python
name = []
value = []
```

```python
#Slimit for parse JS, refer from : https://github.com/rspivak/slimit
for node in nodevisitor.visit(tree):
    if isinstance(node, ast.Assign):
        left, right = node.left, node.right
        leftvalue = left.value
        rightvalue = right.value
        name.append(leftvalue)
        value.append(rightvalue)
        #print(left.value+","+right.value)
        #print(fields)
```

```python
#Split even in list, refer from : https://tutorialspoint.com/How-do-you-split-a-list-into-evenly-sized-chunks-in-Python 
namelist = [name[i:i+13] for i in range(0,len(name),14)]
```

```python
valuelist = [value[i:i+13] for i in range(0,len(value),14)]
```

```python
valuelist
```

```python
df = pd.DataFrame(valuelist, columns=namelist[0])
```
