```python
import json
import pandas as pd
```

```python
with open('C:/Users/chenyit/Downloads/datagovtw_dataset_20191219 (1).json') as json_file:
    data = json.load(json_file)
```

```python
name=[]
disc=[]
coldisc=[]
link=[]
for i in range(len(data)):
    name.append(data[i]['資料集名稱'])
    disc.append(data[i]['資料集描述'])
    coldisc.append(data[i]['主要欄位說明'])
    link.append(data[i]['下載連結'])
```

```python
df = pd.DataFrame({'資料集名稱':name,'資料集描述':disc,'主要欄位說明':coldisc,'下載連結':link},columns=['資料集名稱','資料集描述','主要欄位說明','下載連結'])
```
