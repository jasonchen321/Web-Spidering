```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = requests.get('http://deliciouscatering.hk/shops.html')
url.encoding = 'utf-8'
```

```python
soup = BeautifulSoup(url.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
web = soup.find_all(['p','a'])
```

```python
web
```

```python
infolist=[]
for j in web:
    In = j.string
    if In != None:
        infolist.append(In)
```

```python
info = infolist[5:]
```

```python
name = []
addr = []
phone = []

for i in range(len(info)):
    if i%3 == 0:
        name.append(info[i])
    elif i%3 == 1:
        addr.append(info[i])
    else:
        phone.append(info[i])
```

```python
phonec = [w.replace('電話:','').replace('-','') for w in phone]
```

```python
addrc = [x.replace('地址: ','').replace('地址:','').replace('\n','') for x in addr]
```

```python
df = pd.DataFrame({'Name':name,'Address':addrc,'Phone':phonec},columns=['Name','Address','Phone'])
```
