```python
import requests
import re
import pandas as pd
```

```python
url = requests.get('https://www.mcdonalds.com.hk/googleapps/GoogleHongKongSearchAction.do?method=searchLocation&searchTxtLatlng=(22.25%2C%20114.16669999999999)&actionType=searchRestaurant&country=hk&language=zh')
```

```python
url = url.json()
```

```python
url
```

```python
loclist = url['results']
```

```python
loclist[1]
```

```python
d = loclist[1]['addresses']
```

```python
d
```

```python
name=[]
district=[]
address=[]
tel=[]
lon=[]
lat=[]

for i in range(len(loclist)):
    name.append(loclist[i]['name'])
    district.append(loclist[i]['district'])
    tel.append(loclist[i]['telephone'])
    lon.append(loclist[i]['longitude'])
    lat.append(loclist[i]['latitude'])
    for item in loclist[i]['addresses']:
        addr = item.values()
        for v in addr:
            address.append(v)
```

```python
address
```

```python
df = pd.DataFrame({'name':name,'address':address,'tel':tel,'lontitude':lon,'latitude':lat,'district':district},columns=['name','address','tel','lontitude','latitude','district'])
```

```python
writer = pd.ExcelWriter('C:/Users/chenyit/Desktop/Data.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
