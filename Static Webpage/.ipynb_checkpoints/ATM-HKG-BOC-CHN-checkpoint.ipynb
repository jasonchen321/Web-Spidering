```python
import requests
import re
import pandas as pd
import json
import xmltodict
import xml.etree.ElementTree as ET
import demjson
```

```python
response = requests.get("https://www.bochk.com/services/xml/generatebranchweb.tc.xml?1552530506641")
```

```python
response.encoding ='utf8'
```

```python
r = response.text
```

```python
r_json = json.dumps(xmltodict.parse(r))
rr_json = json.loads(r_json)
```

```python
rr_json['BranchList']['Bank']
```

```python
rr_json['BranchList']['Bank']['Region'][0]['item'][0]['Branch']['Personal']['Address']
```

```python
Name = []
Lat = []
Lon = []
Address = []
atm_Address = []
District = []
Service = []


for i in range(len(rr_json['BranchList']['Bank']['Region'])):
    for j in range(len(rr_json['BranchList']['Bank']['Region'][i]['item'])):
        Name.append(rr_json['BranchList']['Bank']['Region'][i]['item'][j]['Name'])
        Lon.append(rr_json['BranchList']['Bank']['Region'][i]['item'][j]['Longitude'])
        Lat.append(rr_json['BranchList']['Bank']['Region'][i]['item'][j]['Latitude'])    
        District.append(rr_json['BranchList']['Bank']['Region'][i]['Name'])
        try:    
            Address.append(rr_json['BranchList']['Bank']['Region'][i]['item'][j]['Branch']['Personal']['Address'])
        except:
            Address.append("")
        try:    
            atm_Address.append(rr_json['BranchList']['Bank']['Region'][i]['item'][j]['atm']['Address'])
        except:
            atm_Address.append("")
        try:
            Service.append(rr_json['BranchList']['Bank']['Region'][i]['item'][j]['atm']['Services']['Service'])
        except:
            Service.append("")
               
```

```python
df = pd.DataFrame({"Name":Name,"ATM_Addr":atm_Address,"District":District,"X":Lon,"Y":Lat,"Service":Service},
                  columns = ["Name","ATM_Addr","District","X","Y","Service"])
```

```python
#take out rows which ATM_Addr is empty
Only_ATM = df[~(df.ATM_Addr == "")]
```
