```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = requests.get('http://www.taipeimarathon.org.tw/contest.aspx')
soup = BeautifulSoup(url.text,'html.parser')
```

```python
soup
```

```python
table = soup.find_all('table','gridview')
```

```python
table
```

```python
#Get Date
```

```python
date=[]
eventdate = soup.find_all('td','datecell')
for d in eventdate:
    d=d.text
    d=d.replace('\r\n','')
    d=d.replace(' ','')
    date.append(d)
    
```

```python
date
```

```python
len(date)
```

```python
#Get Name of Marathon
```

```python
allevent = soup.find_all('td',{'width':'340'})
```

```python
event=[]
for a in allevent:
    a=a.text
    a=a.replace(' ','')
    a=a.replace('\n','')
    a=a.replace('\r','')
    event.append(a)
```

```python
len(event)
```

```python
df = pd.DataFrame({'Name':event,'Date':date},columns=['Date','Name'])
```

```python
writer = pd.ExcelWriter('C:/Users/chenyit/Desktop/2020 Marathon.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
