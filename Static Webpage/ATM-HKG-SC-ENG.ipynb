```python
import requests
import re
import urllib.request, json
import pandas as pd
```

```python
url = ("https://www.sc.com/hk/data/atm-branch/v2/all-atms-branches.json")
```

```python
web = requests.get(url)
data = json.loads(web.text)
#Method 2
#response = urllib.request.urlopen(url)
#data = json.loads(response.read())
```

```python
name = []
address = []
phone = []
X = []
Y = []

for i in range(len(data['locations'])):
    n = data['locations'][i]['name']
    a = data['locations'][i]['address']['addressLines']
    p = data['locations'][i]['telephone']
    cy = data['locations'][i]['address']['latitude']
    cx = data['locations'][i]['address']['longitude']
    name.append(n)
    address.append(a)
    phone.append(p)
    X.append(cx)
    Y.append(cy)
```

```python
df = pd.DataFrame({'Name':name,'Address':address,'Phone':phone,'Lon':X,'Lat':Y},columns=['Name','Address','Phone','Lon','Lat'])
```
