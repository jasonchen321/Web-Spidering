```python
import requests
url='https://data.gov.hk/assets/data/geospatial/1007979e-552c-4fe3-b3fa-7d72b88396cf.json'
r = requests.get(url) 
```

```python
import pandas as pd
import json
data=eval(r.text)
output=[]
for item in data:
    output.append({'id':item[0],'y':item[1],'x':item[2]})
df=pd.DataFrame(output)
df.head()
```

```python
null=''
name=[]
address=[]

for idx,row in df.iterrows():
    print
    url='https://data.gov.hk/tools/cookiecutter/api/v1/tc/resources/1007979e-552c-4fe3-b3fa-7d72b88396cf/records/'+row['id']
    r = requests.get(url) 
    obj=eval(r.text)
    name.append(obj['results'][0]['rows'][0][3])
    address.append(obj['results'][0]['rows'][0][4])

    
```

```python
df['name']=name
df['address']=address
```
