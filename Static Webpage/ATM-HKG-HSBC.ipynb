```python
# step2. json parsing
all_hsbc=[]
for item in jdata["log"]['entries']:
    if item['response']['content']['mimeType']=='application/json':
        try:
            njson=json.loads(item['response']['content']['text'])
            nr=njson['results']
            
            for pt in nr:
                temp_hsbc=[]                
                temp_hsbc.append(pt['name'])   
                temp_hsbc.append(pt['address']['line1']) 
                temp_hsbc.append(pt['address']['line2']) 
                temp_hsbc.append(pt['address']['line3']) 
                temp_hsbc.append(pt['address']['line4']) 
                temp_hsbc.append(pt['address']['city']) 
                temp_hsbc.append(pt['address']['postalCde']) 
                temp_hsbc.append(pt['address']['ctryCde']) 
                temp_hsbc.append(pt['address']['lat']) 
                temp_hsbc.append(pt['address']['lng']) 
                all_hsbc.append(temp_hsbc) 
        except:
            pass
```

```python
# step3. export to excel
import pandas as pd
df = pd.DataFrame(data=all_hsbc)
df.columns = ['name', 'address1', "address2", "address3", "address4", "city", "postalCde", "ctryCde", "lat", "lng"]
result = df.sort_values('name')
result = result.drop_duplicates()
writer = pd.ExcelWriter('HKG_HSBC_ATM.xlsx')
result.to_excel(writer, 'Sheet1')
writer.save()
result
```
