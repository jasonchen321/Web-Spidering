```python
import xml.etree.ElementTree as ET
from urllib.request import urlopen
import re
import pandas as pd
```

```python
xml = urlopen("http://www.asia.ccb.com/hongkong_tc/personal/map_note.xml")
```

```python
tree = ET.parse(xml)
root = tree.getroot()
```

```python
name = []
address = []
X = []
Y = []
serv = []
for child in root:
    for i in range(3):
        if child.tag == 'title'+str(i+1):
            name.append(child.text)
        if child.tag == 'address'+str(i+1):
            address.append(child.text) 
        if child.tag == 'pointLeft'+str(i+1):
            X.append(child.text)
        if child.tag == 'pointRight'+str(i+1):
            Y.append(child.text)
        if child.tag == 'service'+str(i+1):
            serv.append(child.text)
```

```python
#需再增加Service代碼的替換功能, ATM=condition_2[2]
```

```python
df = pd.DataFrame({'Name':name,'Address':address,'Lon':X,'Lat':Y,'Service':serv},columns=['Name','Address','Lon','Lat','Service'])
```
