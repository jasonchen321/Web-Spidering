```python
import requests
import re
import pandas as pd
```

```python
html = requests.get("https://www.hkbea.com/xml/en/hk-atm-list.txt?_=1552619384114")
html.encoding = None
```

```python
doc = html.text
```

```python
doc
```

```python
rep = {"\r\n":"",
       "{":"",
       "\"branches\" : [":"",
       "\" : \"":"|",
       "\",\"":"|",
       "\"":""
      }
```

```python
for i,j in rep.items():
    doc=doc.replace(i,j)
```

```python
doc
```

```python
ListinList1 = [i.split('|') for i in doc[0:-1].split('},')]
```

```python
ListinList1
```

```python
df = pd.DataFrame(ListinList1)
```
