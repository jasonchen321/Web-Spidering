```python
import requests
import re
import pandas as pd
import json
```

```python
html = requests.get("https://www.hkbea.com/xml/tc/hk-atm-list.txt?_=1552632953209")
doc = html.json()
```

```python
Name = []
Address = []
District = []
Opening = []
Lon = []
Lat = []

for i in range(len(doc['branches'])):
    Name.append(doc['branches'][i]["Name"])
    Address.append(doc['branches'][i]["Addr"])
    District.append(doc['branches'][i]["District"])
    Opening.append(doc['branches'][i]["Remark"])
    Lon.append(doc['branches'][i]["lng"])
    Lat.append(doc['branches'][i]["lat"])

```

```python
df = pd.DataFrame({"Name":Name,"Address":Address,"District":District,"Opening":Opening,"X":Lon,"Y":Lat},
                 columns = ["Name","Address","District","Opening","X","Y"])
```
