```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = requests.get('https://www.parknshop.com/zh-hk/storeLocator')
url.encoding = 'utf-8'
```

```python
soup = BeautifulSoup(url.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
detail = soup.find_all('div',"location-info hidden-content")
shop = soup.find_all('p', "shop-name")
add = soup.find_all('p', "address")
tel = soup.find_all('div', "phone")
openning = soup.find_all('p', "opening-hour")
```

```python
BN = []
X = []
Y =[]
Name = []
Add = []
Tel = []
Open = []
for i in detail:
    BN.append(i.get('data-brandname'))
    X.append(i.get('data-longitude'))
    Y.append(i.get('data-latitude'))
for j in shop:
    Name.append(j.text)
for k in add:
    Add.append(k.text)
for l in tel:
    T = l.text
    T = T.replace("\n","")
    Tel.append(T)
for m in openning:
    M = m.text
    M = M.replace("\r\n\t\t\t\t\t\t\t\t\t\t\t\t\t","")
    Open.append(M)
```

```python
df = pd.DataFrame({"BN":BN,"Name":Name,"Address":Add,"Tel":Tel,"Lon":X,"Lat":Y,"Openning":Open},columns=["BN","Name","Address","Tel","Lon","Lat","Openning"])
```
