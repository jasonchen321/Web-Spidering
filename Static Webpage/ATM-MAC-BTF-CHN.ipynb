```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = requests.get("http://www.taifungbank.com/zh_TW/bank-address")
```

```python
soup = BeautifulSoup(url.text, 'html.parser')
```

```python
soup
```

```python
s = str(soup)
ss = s.find("//$.each(data.posts, function(i, post) {")
#print(ss)
e = s.find("// rabi 20110416")
#print(e)
```

```python
cleandoc = s.replace(s[0:ss+len("//$.each(data.posts, function(i, post) {")],"")
cleandoc = cleandoc.replace(s[e+len("// rabi 20110416"):],"")
```

```python
cleandoc = re.sub(r'<[^<>]*>','', cleandoc)
```

```python
rep = {'\r':'','\n':'','\t':'',')':'',
       '    point = new GLatLng(':'',
       ' = ':'',' += ':'','\"<dt>':'','\"<dd>':'','+':'',
       '\"':''}
for a,b in rep.items():
    cleandoc = cleandoc.replace(a,b)
```

```python
repindex = {'location[':'','messages[':'','lat[':'','lng[':''}
```

```python
for i in range(50):
    for c,d in repindex.items():
        cleandoc = cleandoc.replace(c+str(i)+']',d)
```

```python
d = cleandoc.split(';')
```

```python
doclist = [d[i:i+12] for i in range(0,len(d),12)]
```

```python
del doclist[-1]
```

```python
name = []
address = []
X = []
Y = []

for j in range(len(doclist)):
    name.append(doclist[j][6])
    address.append(doclist[j][4])
    X.append(doclist[j][10])
    Y.append(doclist[j][11])
    
```

```python
df = pd.DataFrame({'Name':name,'Address':address,'X':X,'Y':Y},columns=['Name','Address','X','Y'])
```
