```python
import requests
import re
import pandas as pd
```

```python
html = requests.get("https://www.bochk.com/services/xml/generatebranchweb.en.xml?1552472747020")
```

```python
doc = html.text
```

```python
print(doc)
```

```python
newdoc = doc.replace("<?xml version=\"1.0\" encoding=\"utf-8\" ?><BranchList><Bank><Region><Code><![CDATA[_central_western_district]]></Code><Name><![CDATA[Central &amp; Western District]]></Name><DistrictCode><![CDATA[hong_kong_district]]></DistrictCode><item>", "")
```

```python
newdoc
```

```python
rep = {"<Name><![CDATA[":"|",
       "]]></Name>":"!",
       "<Latitude><![CDATA[":"",
       "]]></Latitude>":"",
       "<Longitude><![CDATA[":" ",
       "]]></Longitude>":"!",
       "<Phone><![CDATA[+":"!",
       "]]></Phone>":"!",
       "<Address><![CDATA[":"",
       "]]></Address>":"@"}
```

```python
for x,y in rep.items():
    newdoc = newdoc.replace(x,y)
```

```python
newdoc
```

```python
cleandoc = re.sub(r'<[^<>]*>','', newdoc) #take time to figure out
```

```python
cleandoc
```

```python
pdoc = cleandoc.split("|")
```

```python
pdoc
```
