MAC_Starbucks_Crawler
==
> Semi-automatic crawler
> v1.0: 2018-07-23

```python
# step1. change 'harfile' as your HAR file dirctory
import json
with open('www.starbucks.com.hk.har', encoding='utf8') as f:
    jdata = json.load(f)
```

```python
# step2. json parsing
all_Starbucks=[]
for item in jdata["log"]['entries']:
    if item['response']['content']['mimeType']=='application/javascript':
        try:              
            st=item['response']['content']['text'].replace("jQuery172003655176013061445_1537179345117(","")  
            st = st[:-1]            
            
            njson=json.loads(st)            
            nr=njson['stores']
            for pt in nr:
                pt=pt['store']
                temp_Starbucks=[]                
                temp_Starbucks.append(pt['name']) 
                temp_Starbucks.append(pt['brandName']) 
                temp_Starbucks.append(pt['address']['city']) 
                temp_Starbucks.append(pt['address']['streetAddressLine1']) 
                temp_Starbucks.append(pt['address']['streetAddressLine2']) 
                temp_Starbucks.append(pt['address']['streetAddressLine3']) 
                temp_Starbucks.append(pt['coordinates']['latitude']) 
                temp_Starbucks.append(pt['coordinates']['longitude']) 
                temp_Starbucks.append(pt['phoneNumber']) 
                all_Starbucks.append(temp_Starbucks) 
        except:
            pass
```

```python
# step3. export to excel
import pandas as pd
df = pd.DataFrame(data=all_Starbucks)
df.columns = ['name', 'brandName', "city", "streetAddressLine1", "streetAddressLine2", "streetAddressLine3", "latitude", "longitude","phoneNumber"]
result = df.sort_values('name')
result = result.drop_duplicates()
writer = pd.ExcelWriter('MAC_Starbucks.xlsx')
result.to_excel(writer, 'Sheet1')
writer.save()
result
```
