```python
import requests
import re
import urllib.request, json
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = ("https://www.fubonbank.com.hk/json/branch_locator_tc.json")
```

```python
web = requests.get(url)
data = json.loads(web.text)
#Method 2
#response = urllib.request.urlopen(url)
#data = json.loads(response.read())
```

```python
name = []
address = []
categ = []
X = []
Y = []
popup = []
phone = []
for i in range(len(data['pins'])):
    n = data['pins'][i]['name']
    a = data['pins'][i]['address']
    c = data['pins'][i]['cat']
    cy = data['pins'][i]['lat']
    cx = data['pins'][i]['lng']
    tel = data['pins'][i]['popupDetails']
    name.append(n)
    address.append(a)
    categ.append(c)
    X.append(cx)
    Y.append(cy)
    #t.append(tel)
    
    #Get tel
    soup = BeautifulSoup(tel, 'html.parser')
    #print(soup.prettify())
    #print("soup.................................................................................................................")
    #print(tel)
    #find the value of tag p
    p_tags = soup.find_all('p')
    #print(p_tags)
    #print("p_tags.................................................................................................................")
    for value in p_tags:
        valueBs4 = value.string
        valuestr = str(valueBs4)
        #print(valuestr)
        #print("valuestr.................................................................................................................")
        popup.append(valuestr)
    #print(popup)
    #print(".................................................................................................................")
    phone.append(popup[5])

        
```

```python
df = pd.DataFrame({'Name':name,'Address':address,'Category':categ,'Phone':phone,'Lon':X,'Lat':Y},columns=['Name','Address','Category','Phone','Lon','Lat'])
```
