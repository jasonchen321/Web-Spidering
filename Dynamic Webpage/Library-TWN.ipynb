```python
from selenium import webdriver
import pandas as pd
```

```python
driver_path = "C:/chromedriver.exe" # 改為你的 driver 路徑
driver = webdriver.Chrome(executable_path = driver_path)

Name = []
Content = []
library_page = 59
for i in range(1,library_page):
    driver.get("http://libstat.ncl.edu.tw/statFront/fairhouse/fsearch.jsp?a=libtype&page="+str(i)+"&keyword=CE")
    library_list = driver.find_elements_by_css_selector(".floatrightbut4+ a")
    for j in range(len(library_list)):
        driver.get("http://libstat.ncl.edu.tw/statFront/fairhouse/fsearch.jsp?a=libtype&page="+str(i)+"&keyword=CE")
        library_list = driver.find_elements_by_css_selector(".floatrightbut4+ a")
        library_list[j].click()
        driver.switch_to.alert.accept()
        library_name = driver.find_elements_by_css_selector("h2")
        library_name = [x.text for x in library_name]
        library_content = driver.find_elements_by_css_selector(".list2 li")
        library_content = [y.text for y in library_content]
        Name.append(library_name)
        Content.append(library_content)
        driver.back()
```

```python
df = pd.DataFrame({"Name":Name,"Text":Content},columns=["Name","Text"])
```
