```python
from selenium import webdriver
import time
driver_path = "C:/chromedriver.exe" # 改為你的 driver 路徑
driver = webdriver.Chrome(executable_path = driver_path)
driver.get("http://zh.macaotourism.gov.mo/dining/restaurant_list.php?dish=")
#res_info = driver.find_elements_by_css_selector(".content_font")
res_click = driver.find_elements_by_css_selector(".travelagen_page a")
```

```python
res = []
for i in range(len(res_click)-1):
    ii = i+1
    driver.get("http://zh.macaotourism.gov.mo/dining/restaurant_list.php?p="+str(ii))
    res_info = driver.find_elements_by_css_selector(".content_font")
    for j in res_info:
        text = j.text
        t = text.split("\n")
        res.append(t)
   
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(res,columns={"Name","Address","Phone"})
```
