```python
import pandas as pd
```

```python
from selenium import webdriver
import time
driver_path = "C:/chromedriver.exe" # 改為你的 driver 路徑
driver = webdriver.Chrome(executable_path = driver_path)
driver.get("https://www.lcsd.gov.hk/tc/facilities/facilitieslist/districts.php?ftid=0")
search_sc = driver.find_elements_by_css_selector(".linkgp_links a")
length = len(search_sc)

sc_n = []
sc_p = []

for i in range(length):
    search_sc = driver.find_elements_by_css_selector(".linkgp_links a")
    search_scs = search_sc[i]
    search_scs.click()
    
    sc_name = driver.find_elements_by_css_selector("h2")
    for a in sc_name:
        a_name = a.text
        sc_n.append(a_name)    
    
    sc_phone = driver.find_elements_by_css_selector("tr:nth-child(2) th+ td") #tr:nth-child(2) th+ td
    for b in sc_phone:
        b_name = b.text
        sc_p.append(b_name)
        
    driver.back()
```

```python
print(sc_n)
```

```python
print(sc_p)
```

```python
a = pd.DataFrame(columns = ['名稱','電話'])
```

```python
a['名稱']=sc_n
```

```python
a['電話']=sc_p
```
